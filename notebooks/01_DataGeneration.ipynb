```python
# Import libraries
import pandas as pd
import numpy as np
from faker import Faker
import random
```

```python
# Initialize Faker
fake = Faker()
```

```python
# Parameters
num_players = 50000
regions = ['North America', 'Europe', 'Asia', 'South America']
devices = ['iOS', 'Android', 'Web']
spend_levels = ['Low', 'Medium', 'High', 'VIP']
```

```python
# Generate players table
players = pd.DataFrame({
    'player_id': range(1, num_players + 1),
    'region': [random.choice(regions) for _ in range(num_players)],
    'device_type': [random.choice(devices) for _ in range(num_players)],
    'join_date': [fake.date_between(start_date='-2y', end_date='today') for _ in range(num_players)],
    'spend_level': [random.choices(spend_levels, weights=[0.5, 0.3, 0.15, 0.05])[0] for _ in range(num_players)]
})
```

```python
# Generate sessions table
sessions = []
for pid in players['player_id']:
    join_date = players.loc[players['player_id'] == pid, 'join_date'].values[0]
    for _ in range(random.randint(5, 50)):
        sessions.append({
            'player_id': pid,
            'session_date': fake.date_between(start_date=join_date, end_date='today'),
            'duration_minutes': random.randint(5, 120),
            'purchases': random.randint(0, 5)
        })

sessions = pd.DataFrame(sessions)
```

```python
# Save to CSV
players.to_csv('data/players.csv', index=False)
sessions.to_csv('data/sessions.csv', index=False)

print("✅ Data generated successfully!")
print(f"Players: {players.shape[0]} rows")
print(f"Sessions: {sessions.shape[0]} rows")
```
